```python
skb=SelectKBest(score_func=chi2,k=10)
```

```python
X_best_features=skb.fit_transform(independent_features,Y)
```

```python
X_best_features
```

```python

# Get the scores of all features
feature_scores = skb.scores_

# Get the names of the selected features
selected_feature_names = independent_features.columns[skb.get_support()]

# Create a DataFrame for better visualization
features_with_scores = pd.DataFrame({
    'Feature': independent_features.columns,
    'Score': feature_scores
})

# Sort by score in descending order
features_with_scores = features_with_scores.sort_values(by='Score', ascending=False)

# Display selected features and their scores
print("Selected Features and Scores:")
print(features_with_scores[features_with_scores['Feature'].isin(selected_feature_names)])
```
